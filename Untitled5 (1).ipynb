```python
import pandas as pd
```

```python
data=pd.read_csv("laptop_data_cleaned.csv")
data
```

```python
# Rata-rata harga berdasarkan Tipe Laptop
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
avg_price_by_type = data.groupby('TypeName')['Price_euros'].mean().sort_values()
avg_price_by_type.plot(kind='barh', color='skyblue')
plt.title('Rata-rata Harga Laptop Berdasarkan Tipe')
plt.xlabel('Harga Rata-rata (Euros)')
plt.ylabel('Tipe Laptop')
plt.grid(axis='x')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur gaya visual (misalnya menggunakan tema seaborn yang lebih ramah)
sns.set_style("whitegrid")

# Mengatur ukuran figur
plt.figure(figsize=(12, 6))

# Membuat visualisasi rata-rata harga berdasarkan tipe laptop
avg_price_by_type = data.groupby('TypeName')['Price_euros'].mean().sort_values()

# Menggunakan skema warna yang lebih hidup dan berwarna-warni untuk memberikan tone yang ramah
colors = sns.color_palette("pastel", len(avg_price_by_type))

# Membuat bar chart horizontal dengan warna khusus dan label yang lebih besar
avg_price_by_type.plot(kind='barh', color=colors)
plt.title('Rata-rata Harga Laptop Berdasarkan Tipe', fontsize=18, fontweight='bold')
plt.xlabel('Harga Rata-rata (Euros)', fontsize=14)
plt.ylabel('Tipe Laptop', fontsize=14)

# Menambahkan anotasi di setiap bar untuk menunjukkan nilai harga
for index, value in enumerate(avg_price_by_type):
    plt.text(value, index, f'{value:.2f}', va='center', ha='left', fontsize=12, color='black', weight='bold')

# Mengatur warna dan gaya grid
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Menampilkan plot
plt.show()
```
