visualisasi diatas dikatakn baik karena:
1. Strive forms and Functions==>
   pada visualisasi diatas,form dan fungsinya berkesinambungan yang dimana form tersebut memberikan informasi terkait dengan data penduduk Sumatera Barat untuk setiap kota pada daerah povinsi tersebut dengan menggunakan diagram batang dan setiap diagram memberikan informasi untuk tiap kabupaten
2. Justifying the selection of everithing we do==>
   justifikasi dari pembuatan visualisasi ini cukup jelas dengan adanya alasan yang diberikan dalam pembuatan visualisasi ini karena menampilkan dengan jelas perbandingan dari setiap penduduk kabupaten melalui diagram batang
3. Creating accessibility through intuitive design==>
   visualisai ini dapat dengan mudah dipahami oleh pembaca dan pengguna dari visualisasi karena pembuatan warna dari setiap diagram memberikan penjelasan terkait perbedaan kepadatan penduduk di setiap kabupaten. serta diberikan keterangan yang menjelaskan hubungan dari warna-warna yang diberikan terhadap nama-nama kabupaten pada provinsi Sumatera Barat tersebut
4. NEVER DECEIVE THE RECEIVER==>
   Penyampaian informasi melalui visualisasi tersebut cukup mudah dipahami  dan memberikan informasi yang relevan terkait dengan skala jumlah penduduk terhadap setiap diagram batangnya.
